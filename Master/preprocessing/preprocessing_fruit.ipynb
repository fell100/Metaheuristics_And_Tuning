```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
```

```python
fruit = pd.read_excel("original_data\Date_Fruit_Datasets\Date_Fruit_Datasets.xlsx")
fruit.info()
```

```python
numeric_columns = fruit.select_dtypes(include=['float64', 'int64']).columns.to_list()
print(numeric_columns)

scaler = StandardScaler()

# Fit and transform only the selected columns
scaled_data = scaler.fit_transform(fruit[numeric_columns])

# Replace the original values in the DataFrame with the scaled values
fruit[numeric_columns] = scaled_data
```

```python
object_columns = fruit.select_dtypes(include=['object']).columns
object_columns
```

```python
label_encoder = LabelEncoder()
fruit[object_columns] = fruit[object_columns].apply(lambda col: label_encoder.fit_transform(col))
```

```python
fruit.to_csv("processed_data/fruit.csv")
```
