```python
def create_excel_view(raw_data, metrica, dataset):
    sample = raw_data[(raw_data['metrica'] == metrica) & (raw_data['Dataset'] == dataset)][['Amostras', 'Pvalue', 'estimate']]
    df = sample

    # Create the matrix DataFrame
    matrix = pd.DataFrame(index=['GA', 'Random Search', 'PSO', 'OPTUNA'], columns=['GA', 'Random Search', 'PSO', 'OPTUNA'])

    # Fill the matrix with arrow symbols and magnitude values
    for index, row in df.iterrows():
        amostra_1, amostra_2 = row['Amostras'].split('_')
        p_value = row['Pvalue']
        magnitude = row['estimate']

        if p_value < 0.01:
            arrow = '●'  # Arrow up
        else:
            arrow = '○'  # Arrow down

        matrix.loc[amostra_1, amostra_2] = f'{arrow} {magnitude:.4f}'
        #matrix.loc[amostra_2, amostra_1] = f'{arrow} {magnitude:.4f}'
        #matrix.loc[amostra_1, amostra_1] = f'-'
        #matrix.loc[amostra_2, amostra_2] = f'-'
        matrix.loc[amostra_2, amostra_1] = f'{arrow} {1 - magnitude:.4f}'  # Fill the reverse relation as well

    it = 0
    for k in range(0,4):
        for j in range(0 + it,4):
            matrix.iloc[k, j] = '-'
            
        it+= 1
    # Print the matrix in a format for Excel
    csv_data = matrix.to_csv(sep='\t')

    print(csv_data)
    #print(matrix)
    return matrix
```
