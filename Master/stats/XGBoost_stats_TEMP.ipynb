```python
import joypy
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Assuming 'df' is your DataFrame containing the data

fig, axes = joypy.joyplot(
    df,
    by='alg_dataset',
    ylim='own',
    hist=False,
    overlap=2,
    linewidth=0.8,
    tails=1.5,
    fade=True,
    grid=True,
    x_range=np.linspace(0.776, 0.98, 100),
    figsize=(9, 6),
    bins=10,
    legend=True
)

# Add quantile lines for each numeric column to each density part
quantiles = [0.25, 0.5, 0.75]  # Add more quantiles if needed
line_colors = ['r', 'g', 'b']  # Specify colors for quantile lines

for ax, data in zip(axes, df.groupby('alg_dataset')):
    for column, color in zip(['F1', 'AUC', 'Accuracy'], line_colors):
        for q in quantiles:
            quantile_value = data[column].quantile(q)
            ax.plot(np.linspace(0.776, 0.98, len(data)), quantile_value, color=color, linestyle='--', linewidth=1.2, label=f'{q} Quantile ({column})')
plt.legend()
plt.show()
```
