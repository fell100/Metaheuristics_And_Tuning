### Teste Mann-Whitney U Test

```python
import pandas as pd


testes = pd.DataFrame(columns=["Dataset", "Amostras","metrica", "U1", "Pvalue", "U2"])
```

```python
from scipy.stats import mannwhitneyu

# Teste entre GA e PSO com a m√©trica F1

#Hipotese nula: as duas distribui√ß√µes n√£o tem diferen√ßa estat√≠stica
#Hip√≥tese alternativa: as duas distribu√ß√µes s√£o diferentes
def realiza_teste(amostra1, amostra2, metrica, dataset_name, str_amostra1, str_amostra2):
    U1, p = mannwhitneyu(amostra1[str(metrica)], amostra2[str(metrica)], method="asymptotic")
    print('=='*22)
    print(f'GA x Optuna - {str(metrica)} score - {dataset_name} Dataset')
    print('=='*22)
    print(f'U1, sendo 1 o algoritmo gen√©tico: {U1}, \np-value correspondente: {p}')

    nx, ny = len(amostra1[str(metrica)]), len(amostra2[str(metrica)])
    U2 = nx*ny - U1
    print(f'U2, sendo 2 optuna: {U2}')
    #temp = pd.DataFrame(columns=["Dataset", "Amostras","metrica", "U1", "Pvalue", "U2"])
    #temp = pd.DataFrame()
    #testes = pd.concat([testes, temp], ignore_index=True)
    temp_dict = {
                'Dataset' :dataset_name, 
                'Amostras' : str_amostra1 + "_" + str_amostra2, 
                'metrica' :metrica, 
                'U1' : U1, 
                'U2' : U2, 
                'Pvalue' : p 
                }
    
    filename = './final_testes_lightGBM.csv'

    try:
        final_df = pd.read_csv(filename)
    except:
        open(filename, "a")
        final_df = pd.DataFrame(columns=["Dataset", "Amostras","metrica", "U1", "Pvalue", "U2"])
    
    temp = pd.DataFrame(temp_dict,index=[0])
    final_df = pd.concat([final_df, temp], ignore_index=True)
    
    final_df[["Dataset", "Amostras","metrica", "U1", "Pvalue", "U2"]].to_csv(filename)
```

```python
import itertools

strings = ["GA", "OPTUNA", "PSO", "Random Search"]

combinations = list(itertools.combinations(strings, 2))

for combination in combinations:
    print(combination)
```

```python
#('GA', 'OPTUNA')
realiza_teste(heart_ga, heart_optuna, 'F1', 'Heart', "GA", "OPTUNA")
realiza_teste(heart_ga, heart_optuna, 'Accuracy', 'Heart', "GA", "OPTUNA")
realiza_teste(heart_ga, heart_optuna, 'AUC', 'Heart', "GA", "OPTUNA")
#('GA', 'PSO')
realiza_teste(heart_ga, heart_pso, 'F1', 'Heart', "GA", "PSO")
realiza_teste(heart_ga, heart_pso, 'Accuracy', 'Heart', "GA", "PSO")
realiza_teste(heart_ga, heart_pso, 'AUC', 'Heart', "GA", "PSO")
#('GA', 'Random Search')
realiza_teste(heart_ga, heart_rs, 'F1', 'Heart', "GA", "Random Search")
realiza_teste(heart_ga, heart_rs, 'Accuracy', 'Heart', "GA", "Random Search")
realiza_teste(heart_ga, heart_rs, 'AUC', 'Heart', "GA", "Random Search")
#('OPTUNA', 'PSO')
realiza_teste(heart_optuna, heart_rs, 'F1', 'Heart', "OPTUNA", "Random Search")
realiza_teste(heart_optuna, heart_rs, 'Accuracy', 'Heart', "OPTUNA", "Random Search")
realiza_teste(heart_optuna, heart_rs, 'AUC', 'Heart', "OPTUNA", "Random Search")
#('OPTUNA', 'Random Search')
realiza_teste(heart_optuna, heart_pso, 'F1', 'Heart', "OPTUNA", "PSO")
realiza_teste(heart_optuna, heart_pso, 'Accuracy', 'Heart', "OPTUNA", "PSO")
realiza_teste(heart_optuna, heart_pso, 'AUC', 'Heart', "OPTUNA", "PSO")
#('PSO', 'Random Search')
realiza_teste(heart_pso, heart_rs, 'F1', 'Heart', "PSO", "Random Search")
realiza_teste(heart_pso, heart_rs, 'Accuracy', 'Heart', "PSO", "Random Search")
realiza_teste(heart_pso, heart_rs, 'AUC', 'Heart', "PSO", "Random Search")
```
